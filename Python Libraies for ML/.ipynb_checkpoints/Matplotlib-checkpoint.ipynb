```python
x = np.linspace(np.pi/4,5*np.pi/4,1000)
y1 = np.sin(x)**2
y2 = np.cos(x)**2
```

```python
plt.plot(x,y1,label='sin')
plt.plot(x,y2,label='cos')
plt.grid()
plt.legend()
plt.show()
```

```python
x = np.arange(-100,100)
fig,ax = plt.subplots(2,2)
ax[0][0].plot(x,x**2,'b')
ax[0][0].set_title('Square')
ax[0][0].label_outer()
ax[0][1].plot(x,x**3,'r')
ax[0][1].set_title('Cube')
ax[0][1].label_outer()
ax[1][0].plot(x,x**4,'b')
ax[1][0].set_title('Power 4')
ax[1][1].plot(x,x,'r')
ax[1][1].set_title('Linear')
ax[1][1].label_outer()
plt.show()
```

```python
from random import randint
x1 = np.array(list(randint(1,100) for _ in range(100)))
x2 = np.array(list(randint(1,100) for _ in range(100)))
```

```python
fig, (ax1,ax2) = plt.subplots(2)
ax1.plot(x1,'r*')
ax2.plot(x2,'b+')
plt.show()
```

```python
plt.plot(x1,label='x')
plt.plot(x2,label='y')
plt.legend()

plt.show()
```

```python
subjects = ['Python', 'C++', 'Ruby', 'Java']
marks = [95,90,80,95]
colors = ['blue', 'brown', 'green', 'red']
explode = (0.1, 0.2, 0, 0)
plt.pie(marks, explode=explode, labels=subjects, colors=colors,autopct='%1.1f%%')
plt.show()
```

```python
x = np.arange(20)
y = x**2
plt.scatter(x,y)
plt.plot(x,y,'black')
plt.show()
```

```python
x = np.arange(20)
y = x**2
fit = np.polyfit(x,y,1)
trend_line = [fit[0]*val for val in x] + fit[1]
plt.scatter(x,y)
plt.plot(trend_line,'r')
plt.plot(x,y,'black')
plt.show()
```
