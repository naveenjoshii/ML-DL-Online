```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.arange(0,10,0.1)
y = np.sin(x)
min_y = [min(y)]*len(y)
max_y = [max(y)]*len(y)
```

```python
# scatter plot
plt.scatter(x,y)
plt.show()
```

```python
# Line plot
plt.plot(x,y)
plt.show()
```

```python
# Giving proper detailing to the graph with labels
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sin Function')
plt.show()
```

```python
plt.plot(x,y,'g*')
plt.grid()
plt.show()
```

```python
# plotting multiple functions in a single plot
plt.plot(x,y,'b-',label='sin(x)')
plt.plot(x, [min(y)]*len(x),'r-',label='max(sin(x))')
plt.plot(x, [max(y)]*len(x), 'r-', label='min(sin(x))')
plt.legend(loc='best')
plt.show()
```

```python
# pie charts
subjects = ['C','Python','Java','DBMS','OS']
marks =  [85,83,90,86,89]
plt.pie(marks, labels=subjects, autopct='%.2f%%')
plt.title('Marks in respective subjects')
plt.show()
```

```python
# bar plots
years = np.array([2015,2016,2017,2018,2019])
sales = np.array([10,20,15,7,17])
plt.bar(years, height=sales)
plt.show()
```

```python
years = np.array([2015,2016,2017,2018,2019])
sales = np.array([10,20,15,7,17])
plt.barh(years, width=sales)
plt.show()
```

```python
# histograms
x = np.random.normal(size=1000)
plt.hist(x)
plt.show()
```

```python
# subplots
years = np.array(['2015','2016','2017','2018','2019'])
sales = np.array([10,20,15,7,17])
investment = np.array([9,15,10,10, 15])
profit = sales - investment
loss = investment - sales

fig, ax = plt.subplots(2,2)
ax[0][0].plot(years, investment)
ax[0][0].set_title('Investment')
ax[0][0].label_outer()
ax[0][1].plot(years, sales)
ax[0][1].set_title('Sales')
ax[0][1].label_outer()
ax[1][0].plot(years, profit)
ax[1][0].set_title('Profit')
ax[1][1].plot(years, loss)
ax[1][1].set_title('Loss')
plt.show()
```
