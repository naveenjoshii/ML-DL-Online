```python
import pandas as pd  #importing pandas library for data preprocessing
```

```python
dataset=pd.read_csv('Training_Dataset_v2.csv')#the dataset is imported
```

```python
dataset.info()#this is used to know no of columns,rows  and the non_null values
```

```python
dataset.drop([1687860],axis=0,inplace=True)#the last row of the dataset is removed because it contains null values
```

```python
dataset.drop('sku',axis=1,inplace=True)#sku doesn't show any effect on the target variable
```

```python
dataset.head(50)
```

```python
dataset['lead_time'].fillna(dataset['lead_time'].mean(), inplace=True)#filling the lead_time column as it contains missing values
```

```python
deck_risk = pd.get_dummies(dataset['deck_risk'],drop_first=True)#geting dummy variables for the categorical varable
```

```python
oe_constraint  = pd.get_dummies(dataset['oe_constraint'],drop_first=True)#getting dummies
```

```python
ppap_risk = pd.get_dummies(dataset['ppap_risk'],drop_first=True)#getting dummies
```

```python
stop_auto_buy = pd.get_dummies(dataset['stop_auto_buy'],drop_first=True)#getting dummies
```

```python
rev_stop = pd.get_dummies(dataset['rev_stop'],drop_first=True)#getting dummies
```

```python
went_on_backorder = pd.get_dummies(dataset['went_on_backorder'],drop_first=True)#getting dummies
```

```python
dataset.head(5)
```

```python
potential_issue = pd.get_dummies(dataset['potential_issue'],drop_first=True)#getting dummies
```

```python
dataset.drop(['potential_issue','deck_risk','oe_constraint','ppap_risk','stop_auto_buy','rev_stop','went_on_backorder'],axis=1,inplace=True)#removing the categorical variables to add the new values
```

```python
dataset.head(5)
```

```python
dataset = pd.concat([dataset,potential_issue,deck_risk,oe_constraint,ppap_risk,stop_auto_buy,rev_stop,went_on_backorder],axis=1)#concatinating the new variables with the dataset  
```

```python
dataset.head(5)
```

```python
#creating matrices for independent variables  and dependent variable
X=dataset.iloc[:, :-1].values#takes all the rows and colums,except last column
Y=dataset.iloc[:, -1].values#takes the last column
```

```python
from sklearn.model_selection import  train_test_split#this is used to know whether our model is overfitting or not
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression#this is a classification problem so i used logistic regression
classifier=LogisticRegression()
classifier.fit(X_train,Y_train)
```

```python
Y_pred=classifier.predict(X_test)#predicting the test set
```

```python
from sklearn.metrics import confusion_matrix#to test the accuracy on test set
cm=confusion_matrix(Y_test,Y_pred)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score#to know the accuracy of training set and test set
```

```python
trainingaccuracy=accuracy_score(Y_train,classifier.predict(X_train))
```
